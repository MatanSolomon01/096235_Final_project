# Temp representation

```python
import pickle as pkl
import pandas as pd
import numpy as np
```

```python
path2root = '..'
```

```python
with open(f'{path2root}/data/data_df.pkl', 'rb') as f:
    data = pkl.load(f)
data
```

```python
with open(f'{path2root}/data/initial_features/temperature/hot_words.txt', 'r') as f:
    HOT_WORDS = f.read().split('\n')
with open(f'{path2root}/data/initial_features/temperature/cold_words.txt', 'r') as f:
    COLD_WORDS = f.read().split('\n')
HOT_WORDS,COLD_WORDS
```

```python
def listO_instructions2_representation(listO_instructions):
    rep = {'cold': 0, 'natural': 0, 'hot': 0}
    last_hot = 0
    last_cold = 0
    for i, ins in enumerate(listO_instructions):
        ins = ins.lower()
        is_hot = any([word in ins for word in HOT_WORDS])
        is_cold = any([word in ins for word in COLD_WORDS])
        if is_hot:
            rep['hot'] += 1
            last_hot = i
        if is_cold:
            rep['cold'] += 1
            last_cold = i
        elif not is_hot:
            rep['natural'] += 0.5
    if not (last_hot == last_cold == 0):
        if last_hot>last_cold:
            rep['hot'] += 2
        else:
            rep['cold'] += 2
    scores = np.array(list(rep.values()))
    return scores/sum(scores)
```

```python
s = data.instructions_3.apply(listO_instructions2_representation)
temp_rep = pd.DataFrame.from_dict(dict(zip(s.index, s.values)),orient='index',columns=['cold', 'natural', 'hot'])
temp_rep
```

We intentionally remove all the recipes with 0 in all the entries

```python
temp_rep = temp_rep[(temp_rep != 0).any(axis=1)]
temp_rep
```

```python
pkl.dump(temp_rep, open(f'{path2root}/data/initial_features/rep_temp.pkl', 'wb'))
```
