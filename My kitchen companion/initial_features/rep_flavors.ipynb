# Flavors representation

```python
import numpy as np
import pandas as pd
import pickle as pkl
from scipy.stats.mstats import gmean
from scipy.stats import hmean
```

```python
FLAVORS = ['sweet', 'salty', 'sour', 'spicy', 'other']
```

```python
path2root = '..'
```

```python
with open(f'{path2root}/data/data_df.pkl', 'rb') as f:
    data = pkl.load(f)
data
```

```python
ing_flavs = pkl.load(open(f'{path2root}/data/initial_features/flavors/ingredients_flavor.pkl', 'rb'))
ing_flavs
```

```python
no_space_lower = lambda list: [x.replace(' ', '').lower() for x in list]
data['ingredients_2'] = data['ingredients_2'].apply(no_space_lower)
data
```

```python
def form_ins_to_flavs_prec(listOings):
    flavs = [ing_flavs[ing] for ing in listOings if ing in ing_flavs.keys()]
    if len(flavs) > 0:
        return [sum([flavor == flav for flavor in flavs]) / len(flavs) for flav in FLAVORS]
    else:
        return [0] * len(FLAVORS)
```

```python
s = data['ingredients_2'].apply(form_ins_to_flavs_prec)
flavs_rep = pd.DataFrame.from_dict(dict(zip(s.index, s.values)), orient='index', columns=FLAVORS)
flavs_rep
```

We intentionally remove all the recipes with 0 in all the entries

```python
flavs_rep = flavs_rep[(flavs_rep != 0).any(axis=1)]
flavs_rep
```

```python
pkl.dump(flavs_rep, open(f'{path2root}/data/initial_features/rep_flavors.pkl', 'wb'))
```
